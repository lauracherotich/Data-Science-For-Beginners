## **Indexing**
### *1D- Vector arrays - take one position*
### *2D- Matric arrays - takes two positions ( row, column)*

## **`Summarizing Data`**

### Aggreaggation methods

# sum(), min(), max, cumsum(), mean()

```python
import numpy as np
client_sales=np.array([[10,20,15],[12,24,36],[17,18,19],[21,22,23],[13,14,15]])
print(client_sales)
```

```python
# sum(axis=0(columns)/1(rows), keepdims=bool)

client_sales.sum(axis=1)# adding values row-wise
client_sales.sum(axis=1)
```

```python
client_sales.sum(axis=0)# adding values column-wise
sum_client_sales=client_sales.sum(axis=1,keepdims=True)
print(sum_client_sales)
new_array=np.concatenate((client_sales, sum_client_sales),axis=1)
print(new_array)
```

```python
print(client_sales.min(axis=1,keepdims=True))# row-wise
print(client_sales.max(axis=0))# column-wise
print(client_sales.mean(axis=1))
```

```python
print(client_sales)
row_wise_cumsum=client_sales.cumsum(axis=1)
print(row_wise_cumsum)
column_wise_cumsum=client_sales.cumsum(axis=0)
print(column_wise_cumsum)
```

# Graphing summary values

```python
client_sales1=np.array([[0,5,1],[5,7,8],[1,3,4],[7,3,4]])
from matplotlib import pyplot as plt
cum_sum_by_client= client_sales1.cumsum(axis=0)
# Fix: The x-axis should have the same number of elements as the y-axis data.
# cum_sum_by_client has 4 rows, so np.arange should also produce 4 values.
x_values = np.arange(cum_sum_by_client.shape[0]) # This will be [0, 1, 2, 3]
plt.plot(x_values, cum_sum_by_client[:,0], label="Client")
plt.plot(x_values, cum_sum_by_client.mean(axis=1,keepdims=True), label='Average')

plt.legend()
plt.show()
```
